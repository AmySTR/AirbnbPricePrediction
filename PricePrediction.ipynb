# Select KBest

```python
#feature selection uses SelectKBest
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2, mutual_info_regression

#Load dataset
df = pd.read_csv('df.csv')
df.logP.value_counts()

#Define X and Y data
X=df.loc[:,df.columns != target]
Y =df.loc[:,df.columns == target]
X.shape
Y.shape

#select the features using SelectKBest and chi square
#select_feature = SelectKBest(score_func=chi2, k=4).fit(X, y)

#select feature with SelectKBest - mutual info regression
select_feature = SelectKBest(score_func=mutual_info_regression, k=15).fit(X, Y)

#Display selected features
selected_features_df = pd.DataFrame({'Feature':list(X.columns), 'Scores': select_feature.scores_})
qw=selected_features_df.sort_values(by='Scores',ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.barh(list(X.columns), sorted(select_feature.scores_), color='maroon', align='center')
    
plt.xlabel("Scores")
plt.ylabel("Features")
plt.title("SelectKBest Feature -mutual info regression")
plt.xticks(rotation =90)
plt.show()
```

# Lasso Regression - Feature selection

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV

base_model = LassoCV(cv=5, random_state=42, max_iter=10000)

# Fit model
model.fit(X,y)
print(model.alpha_)

lasso_best = Lasso(alpha=model.alpha_)
lasso_best.fit(X,y)

print(list(zip(lasso_best.coef_, X)))

print('R squared training set', round(lasso_best.score(X,y)*100, 2))

from sklearn.metrics import mean_squared_error
rmse=mean_squared_error(y, lasso_best.predict(X),squared=False)
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y, lasso_best.predict(X))

print('RMSE', round(rmse, 4))
print('MAE', round(mae, 4))
```
